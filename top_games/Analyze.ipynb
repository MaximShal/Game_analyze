```python
import pandas as pd
from IPython.display import display
from datetime import datetime

pd.set_option('display.max_rows', None)
```

```python
game = pd.read_csv('top150_strategy.csv')
game.to_excel('initial_data.xlsx', index=False)
game
```

```python
def parse_date(date):
    if '|' in date:
        dates = date.split('|')
        date = []
        for date_ in dates:
            date_obj = datetime.strptime(date_, "%B %d, %Y")
            date.append(date_obj.year)
        return 2023 - min(date)
    else:
        date_obj = datetime.strptime(date, "%B %d, %Y")
        return 2023 - date_obj.year 
        
```

```python
game['popularity'] = game.apply(lambda row: row['comments'] / parse_date(row['release_date']), axis=1)
game
```

```python
game.quantile(0.9)
```

```python
game_top14 = game.query('popularity > 42') \
                 .reset_index(drop=True)
game_top14['score'] = game_top14.apply(lambda row: (row['metascore'] + row['userscore']) / 2, axis=1)
game_top14['popularity'] = game_top14['popularity'].round(1)
game_top14['total'] = game_top14.apply(lambda row: row['popularity'] + row['score'], axis=1)
result = game_top14.sort_values(by=['total'], ascending=False).head(3)
```

```python
result.to_csv('result.csv', index=False)
```
